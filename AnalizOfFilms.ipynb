```python
columns_to_replace =['country', 'color', 'director_name', 'movie_imdb_link',
                    'actor_2_name', 'genres', 'actor_1_name',
                    'actor_3_name', 'plot_keywords', 'language',
                    'content_rating'
                    ]

for row in columns_to_replace:
    df[row] = df[row].fillna('unknown')

columns_to_replace = df.keys().difference(columns_to_replace)

for row in columns_to_replace:
    df[row] = df[row].fillna(0)

df.isna().sum()
```

Перерасчитаем деньги на актульаное время с учетом инфляции с 1913 по 2017

```python
from bs4 import BeautifulSoup
import requests


def get_real_value(nominal_amt, old_cpi, new_cpi):
    real_value = (nominal_amt * new_cpi) / old_cpi
    return real_value

url = "https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/"

r = requests.get(url)
data = r.text
soup = BeautifulSoup(data, 'html.parser')

table = soup.find('table')
rows = table.tbody.findAll('tr')

years = []
cpis = []

for row in rows:
    year = row.findAll('td')[0].get_text()
    if year.isdigit() and int(year) < 2017:
        years.append(int(year))
        cpis.append(float(row.findAll('td')[13].get_text()))
        
cpi_table = pd.DataFrame ({
    'year' : years,
    'avg_annual_cpi': cpis
})

CPI_2016 = float(cpi_table[cpi_table['year'] == 2016]['avg_annual_cpi'].iloc[0])

real_domestic_gross = []
real_budget_values = []

for index, row in df.iterrows():
    gross = row['gross']
    budget = row['budget']
    year = row['title_year']
    if year == 0:
        real_domestic_gross.append(gross)
        real_budget_values.append(budget)
        continue
    cpi = float(cpi_table[cpi_table['year'] == int(year)]['avg_annual_cpi'].iloc[0])
    
    real_gross = get_real_value(gross, cpi, CPI_2016)
    real_budget = get_real_value(budget, cpi, CPI_2016)
    real_domestic_gross.append(real_gross)
    real_budget_values.append(real_budget)

df['real_domestic_gross'] = real_domestic_gross
df['real_budget'] = real_budget_values

profits = []
roi_vals = []

for index, row in df.iterrows():
    if row['real_domestic_gross'] == 0 or row['real_budget'] == 0:
        profits.append(0)
        roi_vals.append(0)
        continue
    profit = row['real_domestic_gross'] - row['real_budget']
    budget = row['real_budget']
    num = profit - budget
    den = budget
    roi = (num / den) * 100
    profits.append(profit)
    roi_vals.append(roi)
    
df['profit'] = profits
df['roi'] = roi_vals
df.head()
```

ГИПОТЕЗА 1: США потратила фильмы больше всего денег из всех стран.

Чтобы это проверить нужно сделать выборку по стране и потраченному бюджету и вывести первые топ 5 стран и посмотреть, если в них США и какое место оно там занимает.

Для этого получим список всех стран. 

```python
countrys = df['country'].unique()
countrys
```

Напишем функцию, что принимает на вход имя страны, пробегает в цикле по dataframe и на выходе получает суммарный бюджет, который был потрачен страной за все время на создания фильмов.

```python

def budget_country_for_movies(name_country):
    budget_country = df.loc[df['country'] == name_country, ['real_budget']]
    sum_budget= 0
    for index, real_money in budget_country.iterrows():
        sum_budget += float(real_money.iloc[0])
    return sum_budget
```

Далее создаем словарь, в котором ключами будут являться страны, а значениям - рельтатат функции budget_country_for_movies. Затем преобразуем полученный словарь в dataframe и отсортируем по убыванию.

```python

country_sum_budget = {}
for index_countrys in range(len(countrys)):
    country_sum_budget[countrys[index_countrys]] = budget_country_for_movies(countrys[index_countrys])
    

top5_bigger_country_budget = pd.DataFrame.from_dict(country_sum_budget, orient='index', columns=['budget'])['budget'].sort_values(ascending=False)
top5_bigger_country_budget
```

Гипотеза подтвердилась - США является страной, что тратит на фильм больше всех денег.

ГИПОТЕЗА 3: Японские фильмы жанра комедии лучше кореских жанра драма

```python
ganre_drama = []
ganre_comedy = []
genres = []
for row in df['genres']:
    if 'Comedy' in row and 'Drama' in row:
        ganre_comedy.append(1)
        ganre_drama.append(1)
        row = row.replace('Comedy', '')
        row = row.replace('Drama', '')
        genres.append(row)
    elif 'Comedy' in row and 'Drama' not in row:
        ganre_comedy.append(1)
        ganre_drama.append(0)
        genres.append(row.replace('Comedy', ''))
    elif 'Comedy' not in row and 'Drama' in row:
        ganre_comedy.append(0)
        ganre_drama.append(1)
        genres.append(row.replace('Drama', ''))
    elif 'Comedy' not in row and 'Drama' not in row:
        ganre_comedy.append(0)
        ganre_drama.append(0)
        genres.append(row)

df['genres'] = genres
df['Comedy'] = ganre_comedy
df['Drama'] = ganre_drama
```

Напишем функцию удаления ненужных вертикальных линий(чтобы было)

```python
def delete_double_vertikal_line(parametrs):
    list = []
    for row in df[parametrs]:
        if '||' in row[:2]:
            list.append(row[1:])
        elif '||' in row[-2:]:
            list.append(row[:-1])
        elif '|' == row:
            list.append('')
        elif '|' in row[:1]:
            list.append(row[1:])
        else:
            list.append(row.replace('||', '|', 1))
    
    df[parametrs] = list
        
        
```
